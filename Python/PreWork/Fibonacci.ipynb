```python
fibo_cache = {}
def fibonacci(n):
    if n in fibo_cache:
        return fibo_cache[n]
    if n==1:
        return 1
    if n==2:
        return 1
    elif n>2:
        value= fibonacci(n-1) + fibonacci(n-2)
    fibo_cache[n] = value
    return value

for n in range(1,101):
    print(n,":", fibonacci(n))
```
